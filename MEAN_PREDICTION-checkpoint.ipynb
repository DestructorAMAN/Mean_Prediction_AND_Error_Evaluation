```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data=pd.read_csv("Transformed_Housing_Data2.csv")
data.head()
```

```python
data['mean_sales']=data['Sale_Price'].mean()
data['mean_sales'].head()
# .
# .
# this model predicts that every house will be sold at a perticular price 
# (which is the mean of the price), 
# irrespective of the condition and the orher factors.
```

__Visualising the models__

```python
plt.figure(dpi= 100)
k= range(0,len(data))
plt.scatter(k,data['Sale_Price'].sort_values(),color='red',label='Actual Sale Price')
plt.plot(k,data['mean_sales'].sort_values(),color='green',label='mean_price')
plt.xlabel('Fitted Points (Ascending)')
plt.ylabel('Sale Price')
plt.title('Overall Mean')
plt.legend()
```

```python
grade_mean= data.pivot_table(values = 'Sale_Price', columns='Overall Grade',aggfunc=np.mean)
grade_mean
```

```python
# making a new column
data['grade_mean']=0

# for every grade fill its mean price in a new column:
for i in grade_mean.columns:
    data['grade_mean'][data['Overall Grade']==i]=grade_mean[i][0]
    
data['grade_mean'].head()
```

```python
# How does this model will compare to the previous model:
gradewise_list=[]
for i in range(1,11):
    k=data['Sale_Price'][data['Overall Grade']==i]
    gradewise_list.append(k)
```

```python
classwise_list=[]
for i in range(1,11):
    k= data['Sale_Price'][data["Overall Grade"]==i]
    classwise_list.append(k)
```

```python
plt.figure(dpi=120, figsize=(15,9))
###Plotting "Sale_Price gradewise"###
# z variable is for x-axis
z=0
for i in range(1,11):
    #defining x-axis using z
    points = [k for k in range(z, z+len(classwise_list[i-1]))]
    #plotting
    plt.scatter(points, classwise_list[i-1].sort_values(),
               label=('house with overall grade',i),s=4)
    #plotting gradewise mean
    plt.scatter(points,
               [classwise_list[i-1].mean() for q in range(len(classwise_list[i-1]))],
               s=6,color='pink'
               )
    z=max(points)+1
###plotting Overall Mean###
plt.scatter([q for q in range(0,z)],
           data['mean_sales'],
           color='red',
           label='Overall Mean',
           s=6)
plt.xlabel("Fitted points (Ascending)")
plt.ylabel('Sale_Price')
plt.title('Overall Mean')
plt.legend(loc=4)
```

```python
mean_difference=data['mean_sales']-data['Sale_Price']
grade_mean_difference=data['grade_mean']-data['Sale_Price']
```

```python
k=range(0,len(data))
l=[0 for i in range(len(data))]

plt.figure(figsize=(15,6) ,dpi=100)

plt.subplot(1,2,1)
plt.scatter(k,mean_difference,color='red',label='Residuals',s=2)
plt.plot(k,l,color='green',label='mean Regression',linewidth=3)
plt.xlabel('Fitted points')
plt.ylabel('Residuals')
plt.legend()

plt.subplot(1,2,2)
plt.scatter(k,grade_mean_difference,color='blue',label='Residuals',s=2)
plt.plot(k,l,color='green',label='mean Regression',linewidth=3)
plt.xlabel('Fitted points')
plt.ylabel('Residuals')
plt.legend()
# .
# .
# .
# As it can be seen, the green line depicts the perfect model as 
# it means ZERO residuals.
# 2nd plot is closer to the Zero line which means it is more accurate.
# Hence, The mean regression based on a categorical feature 
# will give better results.
```

__MODEL EVALUATION METRICS__

```python
cost= sum(mean_difference)/len(data)
print(round(cost,7))
# .
# .
# We are getting ZERO because some of our errors
# are positive and some are negative, 
# which cancel each other resulting in ZERO.
```

In order to resolve this we use the Absolute Value.

This is Known as 

1.)__MEAN ABSOLUTE ERROR__

Y hat = the predicted values

Y = actual values

n = total number of the data points

the residual or individual error term is given as:

Y hat - Y

```python
Y=data['Sale_Price']
Y_hat1= data['mean_sales']
Y_hat2=data['grade_mean']
n=len(data)
len(Y),len(Y_hat1),len(Y_hat2)
```

```python
# All three are equal which means we are all set to calculate the error.
cost_mean=sum(abs(Y_hat1-Y))/n
cost_mean
```

```python
cost_mean=sum(abs(Y_hat2-Y))/n
cost_mean
```

Therefore model 2 is better than model 1

```python
from sklearn.metrics import mean_absolute_error
cost_mean=mean_absolute_error(Y_hat1,Y)
cost_grade_mean=mean_absolute_error(Y_hat2,Y)
cost_mean, cost_grade_mean
```

__MEAN SQUARE ERROR__

Similar to the MAE with only difference is that we square the difference.

```python
from sklearn.metrics import mean_squared_error
cost_mean= mean=mean_squared_error(Y_hat1,Y)
cost_grade_mean=mean_squared_error(Y_hat2,Y)
cost_mean, cost_grade_mean
```

__ROOT MEAN SQUARE ERROR__

it is the root of the MSE.                                             
There is no library function for RMSE, therefore we will make our own.

```python
from sklearn.metrics import mean_squared_error
cost_mean= mean=mean_squared_error(Y_hat1,Y)**0.5
cost_grade_mean=mean_squared_error(Y_hat2,Y)**0.5
cost_mean, cost_grade_mean
```

__Calculating the accuracy in terms of percetage__

R^2= 1 - ((summ.(i,n) (Y cap - Yi)^2)/                                                                        -------------------------------
          (summ.(i,n) (Y mean - Yi)^2))

Ratio of mean square error of the regression model we want to evaluate 

to the 

Ratio of the mean square error of the simple mean regession model.

0<R^2<1

__>>  It can be negative when the model that we are trying to evaluate is
even worse than the mean regression model__

```python
Y= data['Sale_Price']
Y_bar=data['mean_sales']
Y_hat=data['grade_mean']
n=len(data)
len(Y),len(Y_hat),len(Y_bar),(n)
```

```python
mse_mean=mean_squared_error(Y_bar,Y)
mse_mean
```

```python
mse_model=mean_squared_error(Y_hat,Y)
mse_model
```

```python
R2=1-(mse_model/mse_mean)
print(R2)
print("OR")
print(R2*100)
```
